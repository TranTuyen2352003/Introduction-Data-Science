```python
# Question 8:
def calc_sum(nums):
    pass
    sum = 0
    for x in nums:
        sum += x
    return sum
    # Your code ends here
nums = [5, 3, -2, 4, -1]
calc_sum(nums)
```

```python
#Question 9:
def count_char_freq(string):
    pass
    dict = {}
    for i in string:
        if i in dict:
            dict[i] += 1
        else:
            dict[i] = 1
    return dict
    # Your code ends here
string = "abcaccdefeef"
char_freq = count_char_freq(string)
char_freq
```
